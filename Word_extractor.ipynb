#0. import files from google drive to google colab (optional) 

```
# Google driveにcolabのファイルを保存する。
# https://qiita.com/kakinaguru_zo/items/33dbe24276915124f545

upload_filename = 'dictionary-of-legal-terms.pdf'

file_list = drive_service.files().list(q="name='" + upload_filename + "'").execute().get('files')

# get the file id
file_id = None
for file in file_list:
  if file.get('name') == upload_filename:
    file_id = file.get('id')
    break

if file_id is None:
  print(upload_filename + 'is not found.')
else:
  # upload files to colab
  with open(upload_filename, 'wb') as f:
    request = drive_service.files().get_media(fileId=file_id)
    media = googleapiclient.http.MediaIoBaseDownload(f, request)

    done = False
    while not done:
      progress_status, done = media.next_chunk()
      print(100*progress_status.progress(), end="")
      print("% completed")

  print("completed importing files from google drive to google colab.")
```

#2. extract all the words from a dictionary(.pdf)

```
# path to the dictionary
filename ='/content/drive/My Drive/Colab Notebooks/dictionary-of-legal-terms.pdf' 

pdfFileObj = open(filename,'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)   #The pdfReader variable is a readable object that will be parsed
num_pages = pdfReader.numPages                 #discerning the number of pages will allow us to parse through all the pages


count = 12 # start page that you extract words from
text = ""
                                                            
while count < 626: # end page that you extract words from
    pageObj = pdfReader.getPage(count)
    count +=1
    text += pageObj.extractText()
    
#Below if statement exists to check if the above library returned #words. It's done because PyPDF2 cannot read scanned files.
if text != "":
    text = text 
    
#If the above returns as False, we run the OCR library textract to #convert scanned/image based PDF files into text
else:
    text = textract.process('http://bit.ly/epo_keyword_extraction_document', method='tesseract', language='eng')

    # Now we have a text variable which contains all the text derived from our PDF file.
```

#3. pick up the title words and put them to an external file(dicWords)

```
import pprint
```

```
# extract index words with ragex
capital = r'[A-Z]+[.?(A-Z)+| ?(A-Z)+]+[A-Z.]'
words=re.findall(capital,text)
pprint.pprint(words)
```

```
# check the process
answer=["AA.B.A.","ABANDONED CHILD","ABANDONMENT","ABATABLE NUISANCE","ABATEMENT","ABATEMENT OF A LEGACY","ABATEMENT OF TAXES","ABDUCTION","ABET","ABEYANCE",
"ABILITY TO STAND TRIAL",
"AB INITIO","ABNORMALLY DANGEROUS ACTIVITY","ABOLISH",
"ABORIGINAL TITLE",
"ABORTION",
"ABRIDGE",
"ABROGATE",
"ABSCOND",
"ABSENTIA",
"ABSOLUTE LIABILITY",
"ABSQUE HOC",
"ABSTENTION",
"ABSTRACT OF RECORD"]

def check2():
  if len(answer)==len(words):
    print("yiepee")
  else:
    print("missing words are:")
    pprint.pprint(set(answer)-set(words))
    print("redundant words are:")
    pprint.pprint(set(words)-set(answer))
check2()
```

```
words=set(words) # kill redundant words from the list
dicWords = sorted(words) # sort the words
pprint.pprint(dicWords)

# dicWords is the list of all the index words extracted from the dictionary
# output dicWords to an external file
dicfile=open("/content/drive/My Drive/Colab Notebooks/0_dicWords.txt","w")
for i in range(len(dicWords)):
  dicfile.write(dicWords[i]+"\n")
dicfile.close()
```

#4. extract all the words from each textbook(subjects[i][0]) and put them into subject wordlists(subjects[i][1])

```
# set up function
def clean(i): #i: subjects[i][0]  
  op = open(subjects[i][0],"r") #type(open())=str 
  read=op.read() #type(str.read())=str
  read=re.sub(r"\W"," ", read)
  read=re.sub(r"[0-9]"," ", read)
  read=read.upper() #ここまでtype(read)=str
  read=read.split() #ここで  type(read)=list 
  
  wordset={}
  wordset=set(read) #to remove redundant words
  for ii in wordset:
    subjects[i][1].append(ii) #add cleaned words from wordset #i=index number here!!    
  op.close()
```

```
# run the function to each subject
for i in range(len(subjects)):
  clean(i)

  # check the result
  print(str(i)+"length")
  print(len(subjects[i][1]))
  print(str(i)+"subset")
  print(subjects[i][1][1:10])
```
